```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm
from NormalFunctions import NormalInverseCDF, NormalCDF
```

```python
x = np.arange(0.001, 0.999, 0.001)
y = np.zeros(len(x))
y_true = y
for i, num in enumerate(x):
    y[i] = NormalInverseCDF(x[i])
    y_true[i] = norm.ppf(x[i])
```

```python
plt.plot(x,y,label = 'Moro')
plt.plot(x,y_true,'--', label = 'True')
plt.title('Moro Estimate vs True Value of Inverse Normal CDF')
plt.xlabel('$x$')
plt.ylabel('$N^{-1}(x)$')
plt.legend()
plt.show()
print('Max difference = ', max(abs(y-y_true)))
```

```python
x = np.arange(-2,2,0.01)
y = np.zeros(len(x))
y_true = y
for i, num in enumerate(x):
    y[i] = NormalCDF(x[i])
    y_true[i] = norm.cdf(x[i])

```

```python
plt.plot(x,y,label = 'Our Estimate')
plt.plot(x,y_true,'--', label = 'True')
plt.title('Our Estimate vs True Value of Normal CDF')
plt.xlabel('$x$')
plt.ylabel('$N(x)$')
plt.legend()
plt.show()
print('Max difference = ', max(abs(y-y_true)))
```
